```python
import pandas as pd
import os
import seaborn as sns
import random
```

```python
histSuffix = '/hist.csv'
bdSuffix = '/bd.log'
fitSuffix = '/fit.log'

def strToPoint(s: str):
    return tuple([float(i) for i in s.strip('array([').strip('])').split(',')])

def splitIgnoreBracket(s: str):
    out = []
    cur = ''
    doSplit = True
    for i in s:
        if i == '[':
            doSplit = False
        if i == ']':
            doSplit = True
        if i == ',' and doSplit:
            out.append(cur.strip(' '))
            cur = ''
        else:
            cur += i
    out.append(cur.strip(' '))
    return out

def getArmCoords(interAll, indNumber):
    arm_x = []
    arm_y = []
    for x, y in interAll[indNumber]:
        arm_x.append(x)
        arm_y.append(y)
    return arm_x, arm_y

def plotArm(interAll, indNumber, label=None):
    armx, army = getArmCoords(interAll, indNumber)
    plt.plot(armx, army, label=label)

def show_plot(x, y, plot_gen,n):
    plt.scatter(x,y,alpha = 0.5, label='Total explored space')
    start_gen = plot_gen
    for nb_gen in range(n):
        index = (start_gen + nb_gen) * 200 + 100
        plt.scatter(x[index:index+200],y[index:index+200],label=f'{start_gen+nb_gen}th Generation')
    plt.legend()
    plt.show()

def env_plot():
    wallx = [-3,-1,-1,-3]
    wally = [-1,-1,1,1]
    plt.plot(-2,0,label="target",marker ='x',color='violet')
    plt.plot(wallx,wally, color='black',lw = 3)

def bdLogToData(filepath):
    with open(filepath) as f:
        x = []
        y = []
        for fline in f.readlines():
            fl = fline.strip('\n').split(' ')
            x.append(float(fl[0]))
            y.append(float(fl[1]))
        return x, y

def fitLogToData(filepath, successThreshold = -0.5):
    firstSuccess = -1
    with open(filepath) as f:
        out = {}
        curgen = -1
        curGenFits = []
        for logLine in f.readlines():
            if logLine.startswith("##"):
                linesplit = logLine.split(' ')
                if curgen != -1:
                    out[curgen] = max(curGenFits)
                    if out[curgen] >= successThreshold and firstSuccess == -1:
                        firstSuccess = curgen
                    curGenFits = []
                curgen = int(linesplit[2].strip('\n').strip(' '))
                continue
            curGenFits.append(float(logLine.strip('\n')))
        return out, firstSuccess

def builddb(completedRunsPath, runTypes=None):
    if runTypes is None:
        runTypes = ['FIT_NS', 'NSLC', 'FIT', 'NS']
    runTypes = sorted(runTypes, key=len)
    runTypes.reverse()
    out = {k:[] for k in runTypes}

    for fPath, folderNames, _ in os.walk(completedRunsPath):
        for fName in folderNames:
            curKey = ''
            for t in runTypes:
                if t in fName:
                    curKey = t
                    break
            if curKey == '':
                raise ValueError

            gen = []
            fit = []
            ind = []
            inter = []
            pathPrefix = fPath + '/' + fName + '/'
            with open(pathPrefix + histSuffix) as csvfile:
                for line in csvfile.readlines()[1:]:
                    l = line.strip('\n').split(';')
                    gen.append(int(l[0]))
                    fit.append(float(l[1]))
                    indL = [float(i) for i in l[2].strip("[").strip("]").split(",")]
                    ind.append(indL)
                    interPt = [strToPoint(pt) for pt in splitIgnoreBracket(l[3].strip('[').strip(']'))]
                    inter.append(interPt)

            fgm, firstSuccess = fitLogToData(pathPrefix + fitSuffix)

            entry = {
                'name': fName,
                'gen': gen,
                'fit': fit,
                'ind': ind,
                'inter': inter,
                'bd': bdLogToData(pathPrefix + bdSuffix),
                'fitGenMax': fgm,
                'firstSuccess': firstSuccess
            }
            out[curKey].append(entry)
    return out

def make_stats(database, run_type, pickedRun=0):
    runList = database[run_type]
    firstSuccesses = [run['firstSuccess'] for run in runList]
    successCount = sum([1 for run in runList if run['firstSuccess'] != -1])
    successRate = successCount / len(runList)

    print(f'TYPE: {run_type}\n'
          f'SUCCESS RATE: {successRate} ({successCount} / {len(runList)}) '
          f'FIRST SUCCESS FOR ALL RUNS: {firstSuccesses}\n')

    entries = []
    for i in runList[0]['fitGenMax']:
        for runNb in range(len(runList)):
            entries.append({
                'RunNb': runNb,
                'Generation': i,
                'MaxFit': runList[runNb]['fitGenMax'][i]
            })

    df = pd.DataFrame(entries)

    sns.lineplot(data=df, x='Generation', y='MaxFit')
    plt.axhline(y = -0.5, color = 'r', linestyle = '-')
    plt.title(f'Average fitness over Generations ({run_type})')
    plt.figure()
    ax = sns.lineplot(data=df, x='Generation', y='MaxFit', hue='RunNb')
    sns.move_legend(ax, 'lower right')
    plt.axhline(y = -0.5, color = 'r', linestyle = '-')
    plt.title(f'Max fitness of each runs over Generations ({run_type})')
    plt.figure()

    run_sample = runList[pickedRun]
    plt.plot(run_sample['gen'], run_sample['fit'])
    plt.axhline(y = -0.5, color = 'r', linestyle = '-')
    plt.title('Fitness of selected run')

    plt.figure()

    firstSuccess = firstSuccesses[pickedRun]

    if firstSuccess != -1:
        try:
            argind = run_sample['gen'].index(firstSuccess)
            rng = random.randint(1, argind - 1)
            fs_sample = [run_sample['gen'][rng], firstSuccesses[pickedRun], run_sample['gen'][-1]]
            fs_sample_labels = ['Pre Success Ind.', 'First Success Ind.', 'Last Ind.']
        except ValueError:
            fs_sample = [firstSuccesses[pickedRun], run_sample['gen'][-1]]
            fs_sample_labels = ['First Success Ind.', 'Last Ind.', 'Individu aléatoire']
            rng = random.choice(run_sample['gen'])
            cap = 0
            while rng in fs_sample and cap < 10:
                rng = random.choice(run_sample['gen'])
            fs_sample.append(rng)
    else:
        fs_sample = random.sample(run_sample['gen'], 3)
        fs_sample.sort()
        fs_sample_labels = [None, None, None]

    for i, label in zip(fs_sample, fs_sample_labels):
        plotArm(run_sample['inter'], run_sample['gen'].index(i), label)

    plt.title(f'Arms of run {pickedRun} for generations {fs_sample}')
    plt.legend()
    env_plot()
    plt.show()

    return df
```

## 4 Ajout d'une qualité locale

L'ensemble des solutions générées peut être utilisé pour atteindre n'importe lequel des comportements atteignables, mais l'inconvénient de cette approche est que la notion de qualité est totalement absente du processus, or parmi les solutions générant un comportement donné, toutes ne se valent pas. Certaines sont plus intéressantes que d'autres parce qu'elle consomment moins d'énergie, qu'elles ne créent pas de collision, qu'elles sont plus stables, etc.

Une solution pour prendre en compte un tel critère de qualité consiste à utiliser, à côté de l'objectif de nouveauté, un objectif de performance. Définir cet objectif comme une pression globale est contreproductif, car pour éviter des collisions ou minimiser la consommation d'énergie, il suffit de ne pas bouger... Pour rendre cette pression plus intéressante, il faut en faire un objectif non pas global, mais local.

Pour cela, on peut suivre l'approche proposée par Lehman et Stanley [1]: on compare la fitness de l'individu considéré avec celle de ses plus proches voisins (qui sont déjà déterminés pour le calcul de nouveauté). On ajoute alors un objectif de compétition locale qui vaut le nombre de voisins dont la fitness est inférieure.

Complétez le code de novelty_search.py pour que la fonction de calcul de nouveauté renvoie la nouveauté et l'objectif de compétition locale. Pour cela, vous devrez garder dans l'archive la liste des fitness des points ajoutés.

### 5.2 MAP-Elites

#### 5.2.1 Implémentation de MAP-Elites

La grille définie à la question précédente permet de définir un algorithme très simple: MAP-Elites [1]. Dans cet algorithme, la sélection s'appuie sur la grille. La génération d'un nouvel individu consiste à tirer aléatoirement un (si mutation uniquement) ou deux individus (si croisement) dans la grille puis à appliquer l'opérateur génétique de mutation ou de croisement. Après évaluation, on tente d'ajouter cet individu dans la grille. Si la case correspondante est vide, on l'ajoute, sinon, on le compare à l'individu déjà présent et on le remplace si la fitness est améliorée.
